```
def clear_environment(db_name):
    current_user = (spark
                .createDataFrame([("",)], "user string")
                .withColumn("user", F.expr("current_user"))
                .head()["user"]
               )
    spark.sql(f" DROP DATABASE IF EXISTS {db_name} CASCADE")
    dbutils.fs.rm(f"/FileStore/{current_user}",recurse = True)
    
    get_database().display()
    return(True)

#get all database from env
def get_database():
    database_list = spark.catalog.listDatabases()
    database_rdd = spark.sparkContext.parallelize(database_list)
    database_df = database_rdd.toDF()
    return(database_df)    
```
