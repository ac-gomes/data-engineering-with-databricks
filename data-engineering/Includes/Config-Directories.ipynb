```
from pyspark.sql import functions as F

# Directories that wiil be created.
directory_dict = {
     'main_dir': {
        'isCreated': False,
        'path': '',
    },
    'raw': {
        'isCreated': False,
        'path': '',
    },
    'structured': {
        'isCreated': False,
        'path': '',
    },
    'curated': {
        'isCreated': False,
        'path': '',
    },   
}

#Get current user
current_user = (spark
                .createDataFrame([("",)], "user string")
                .withColumn("user", F.expr("current_user"))
                .head()["user"]
               )


def set_WorkDirectory(directory_dict: dict, current_user: str):
    if len(current_user) > 0:
        user_dir = {item: v for item,
                  v in directory_dict.items() if item.startswith('main')}
        user_dir['main_dir']['isCreated'] = dbutils.fs.mkdirs(f"/FileStore/{current_user}")
        
        if user_dir['main_dir']['isCreated'] == True:
            user_dir['main_dir']['path'] = f"/FileStore/{current_user}"          

            for directory in directory_dict:
                if directory != 'main_dir':
                    directory_dict[directory]['isCreated'] = dbutils.fs.mkdirs(f"/FileStore/{current_user}/{directory}")
                    directory_dict[directory]['path'] = f"/FileStore/{current_user}/{directory}"
        
        isCreated_directories = directory_exists(directory_dict)
    
    return(isCreated_directories)  

def directory_exists(directory_dict: dict) -> bool:
    for dir in directory_dict:
        isCreated = directory_dict[dir]['isCreated'] 
        if isCreated != False:
            return(isCreated)
    return(False)


def get_dirs():   
    try:
        directories = dbutils.fs.ls(f"/FileStore/{current_user}")
        directories = spark.createDataFrame(directories)
        
        if len(directories.columns) > 0:
            directories.display()
            return(True)
    except:
        displayHTML("""
            Something went wrong!
                <li><span style="color:red; font-weight:bold">directory does not exist!</span></li>
        """)
```

```
set_WorkDirectory(directory_dict,current_user)
get_dirs()
```
