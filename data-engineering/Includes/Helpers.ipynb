```
#Test dataframe quality

def get_dataframe_metadata(df: DataFrame):
    metadata = []
    try: 
        for col in df.dtypes: 
            metadata.append(tuple(f"{col[0]},{col[1]}".split(",")))
                            
        return metadata        
 
    except Exception as Error:
        print(f"Something went wrong: {Error}")
        
    return False                    
    
    
```
