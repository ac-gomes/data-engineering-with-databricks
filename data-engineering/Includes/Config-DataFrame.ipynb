```
from pyspark.sql.types import StringType, StructField, StructType

#Setup for generate the dataframes
def create_dataframe(schema: list, data: list):    
    
    if schema != None and data != None:
        schema_df = StructType(schema)
        df1 = spark.createDataFrame(data, schema_df)
        
        return(df1)
    else:
        print(f"Please insert a list with schema and a list with data!")
```
